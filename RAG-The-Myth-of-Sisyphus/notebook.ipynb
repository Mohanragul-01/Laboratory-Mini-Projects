```python
from sentence_transformers import SentenceTransformer, util
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
```

```python
from google.colab import userdata
api_key = userdata.get('GROQ_API_KEY')
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="gemma2-9b-it",
    api_key=api_key,
)
```

```python
with open("/content/myth_of_sisyphus.txt", "r", encoding="utf-8") as f:
    text = f.read()

import re
sentences = re.split(r'(?<=[.!?]) +', text)
chunks = [" ".join(sentences[i:i+4]) for i in range(0, len(sentences), 4)]
chunks
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

model_name = "sentence-transformers/all-MiniLM-L6-v2"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': False}
hf = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
from langchain_chroma import Chroma

vector_store = Chroma(
    collection_name="collection",
    embedding_function=hf,
    persist_directory="chroma_langchain_db",
)
```

```python
from langchain.docstore.document import Document

documents = [Document(page_content=chunk) for chunk in chunks]
vector_store.add_documents(documents)
```

```python
from langchain import hub
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import PromptTemplate


short_response_template = """You are an AI assistant. Respond to the user's question based on the provided context, and keep your answer concise and to the point.
    Context:
    {context}

    Question:
    {question}

    Short Answer:"""

short_response_prompt = PromptTemplate.from_template(short_response_template)

retriever = vector_store.as_retriever()

def retrieve_documents(query: str, retriever):
  return retriever.invoke(query)

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

def generate_response(query: str, retriever, llm, rag_prompt):
    rag_chain = (
        {"context": retriever | format_docs, "question": RunnablePassthrough()}
        | short_response_prompt
        | llm
        | StrOutputParser()
    )
    return rag_chain.invoke(query)

queries = [
    "What is the central message of 'The Myth of Sisyphus'?",
    "How does Camus define the absurd?",
    "Why does Camus consider suicide a fundamental philosophical problem?",
    "How does Camus interpret the myth of Sisyphus?",
    "What role does hope play in Camus's view of the absurd?",
    "How does Camus relate absurdism to existentialism?",
    "What is the relationship between reason and the absurd in Camus's work?",
    "How does Camus suggest one should respond to the absurd?",
    "What does Camus mean when he says 'One must imagine Sisyphus happy'?",
    "How does Camus critique religion and metaphysical beliefs in the essay?",
]

for query in queries:
    print(f"\nQuery: {query}")
    response = generate_response(query, retriever, llm, short_response_prompt)
    print(f"Answer: {response}")
```
